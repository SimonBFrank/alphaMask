```python
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
h, w = [], []
for f in os.listdir('data/images/training/'):
    filename = 'data/images/training/' + f
    img = plt.imread(filename)
    h.append(img.shape[0])
    w.append(img.shape[1])
```

```python
h = np.array(h)
w = np.array(w)
```

```python
print(np.mean(h))
print(np.median(h))
```

```python
sns.violinplot(h)
plt.show()
```

```python
print(np.mean(w))
print(np.median(w))
```

```python
sns.violinplot(w)
plt.show()
```
