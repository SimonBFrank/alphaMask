```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import cv2
import os

%matplotlib inline
```

```python
color_dict = {0:'r', 1:'orange', 2:'g'}
```

```python
img_paths = ['../processed/images/'+path for path in os.listdir('../processed/images/')]
annotation_paths = ['../processed/annotations/'+path for path in os.listdir('../processed/annotations/')]

for img_path, annotation_path in zip(img_paths[:5], annotation_paths[:5]):
    img = np.load(img_path)
    annotation = np.load(annotation_path, allow_pickle=True)
    for bbox in annotation:
        plt.gca().add_patch(Rectangle(*bbox[0] ,linewidth=2,edgecolor=color_dict[bbox[1]],facecolor='none'))
    plt.imshow(img)
    plt.show()
```

```python
class1, class2, class0 = 0, 0, 0
for file in os.listdir('../processed/annotations/'):
    annotation = np.load('../processed/annotations/'+file, allow_pickle=True)
    for bbox in annotation:
        if (bbox[1] == 0):
            class0 += 1
        elif (bbox[1] == 1):
            class1 += 1
        else:
            class2 += 1
print(class0, class1, class2)
```
